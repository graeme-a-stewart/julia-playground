```julia
function monthstats(df, year, month)
    startdate = Date(year, month)
    enddate = Date(year, month, Dates.daysinmonth(startdate))
    # println("$startdate $enddate")
    datemask = (df[!, :DateTime] .>= startdate) .& (bp_df[!, :DateTime] .<= enddate)
    dfview = @view df[datemask, :]
    if nrow(dfview) == 0
        return nothing
    end
    systolic = mean(dfview[!, :Systolic])
    systolic_std = stdm(dfview[!, :Systolic], systolic)
    diastolic = mean(dfview[!, :Diastolic])
    diastolic_std = stdm(dfview[!, :Diastolic], diastolic)
    startdate, systolic, systolic_std, diastolic, diastolic_std
end
```

```julia
struct DateRange
    starty::Int
    startm::Int
    endy::Int
    endm::Int
end
```

```julia
function Base.iterate(d::DateRange, state=0)
    y=d.starty
    m=d.startm
    m+=state
    yo,mo = divrem(m-1, 12)
    m=mo+1
    y+=yo
    if (y > d.endy) || ((y == d.endy) && (m > d.endm))
        return nothing
    end
    return (y, m), state+1
end
```

```julia
bpmonth=DataFrame(date=Date[], sys=Float64[], sysstd=Float64[], dia=Float64[], diastd=Float64[])
for (y,m) in DateRange(2020,6,2023,8)
    println("$y-$m")
    push!(bpmonth, monthstats(bp_df, y, m))
end
```

```julia
@df bpmonth scatter(:date, :sys, label="Systolic BP")
#@df bpmonth plot!(:date, :sys, label="", lc=:blue)
@df bpmonth scatter!(:date, :dia, label="Diastolic BP", mc=:brown)
#@df bpmonth plot!(:date, :dia, label="", lc=:brown)
plot!([minimum(bpmonth[!,:date]), maximum(bpmonth[:,:date])], [80, 80], lc=:red, ls=:dash, label="")
plot!([minimum(bpmonth[!,:date]), maximum(bpmonth[:,:date])], [140, 140], lc=:red, ls=:dash, label="")
for (date, s, ss) in eachrow(bpmonth[!, [:date, :sys, :sysstd]])
    plot!([date, date], [s-ss, s+ss], lc=:blue, label="")
end
for (date, d, ds) in eachrow(bpmonth[!, [:date, :dia, :diastd]])
    plot!([date, date], [d-ds, d+ds], lc=:brown, label="")
end
title!("Blood Pressure")
xlabel!("Date")
ylabel!("mmHg")
```
